```python
import ee

ee.Initialize()
```

```python
from cnwi.modeling import SmileRandomForest

# Load the data
data = ee.FeatureCollection("projects/ee-nwrc-geomatics/assets/cnwi/sampled_features_122")
```

```python
from pprint import pprint
# pprint(data.first().getInfo())

predictors = data.first().getInfo()['properties'].keys()
predictors = [p for p in predictors if p not in ['system:index', 'system:time_start', 'system:time_end', 'system:footprint']]
pprint(predictors)
```

```python
model = SmileRandomForest(
    numberOfTrees=100,
)

model.fit(features=data, classProperty='class_name', inputProperties=predictors)

pprint(model.model.getInfo())
```
